#### Simulate "Bump" in TESS lightcurve

LightCurveSimulator is a class which allows the user to simulate "bumps" in TESS light curves. The initializer takes in either TessLightCurve objects (def in the lightkurve library)
or a path to a lightcurve fits file

In order to generate new simulated lightcurves, the user must call the add_error() method which randomly selects a data point along the lightcurve to add a bump to. The magnitude of this bump is a percentage of the mean flux in the original lightcurve. This percentage is a call parameter automatically set to 0.1 (10%) in the class initializer and can be modified if needed.

The original and new lightcurve can be plotted below with the plot method in the LightCurveSimulator class. The method allows the user to modify the plot title through an optional title parameter

```python
lcsim = LightCurveSimulator(lcs[0])
lcsim.add_error()
lcsim.plot()
```
